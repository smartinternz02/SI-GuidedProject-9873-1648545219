# Activity - 2: Bar Graph

```python
plt.figure(figsize=(10,5))
plt.bar([0.5,1.25,2.25,3.25,4.25],[10,25,35,45,50],label='Iphone')
plt.bar([0.75,1.75,2.75,3.74,4.76],[11,22,33,44,55],label='samsung')
plt.legend()
```

```python
plt.figure(figsize=(10,5))
plt.bar([0.5,1.25,2.25,3.25,4.25],[10,25,35,45,50],label='Iphone',width=0.5)
plt.bar([0.75,1.75,2.75,3.74,4.76],[11,22,33,44,55],label='samsung',width=0.5)
plt.legend()
```

```python
plt.figure(figsize=(10,5))
plt.bar([0.5,1.25,2.25,3.25,4.25],[10,25,35,45,70],label='Iphone',width=0.5)
plt.bar([0.75,1.75,2.75,3.74,4.76],[11,22,30,14,25],label='samsung',width=0.5)
plt.legend()
```

# Activity - 6: Heatmap

```python
corr_matrix=df.corr()
```

```python
sns.heatmap(corr_matrix)
```

```python
plt.figure(1,figsize=(10,5))
```

```python
sns.heatmap(corr_matrix,cmap='BrBG')
```

```python
plt.figure(1,figsize=(10,5))
sns.heatmap(corr_matrix,cmap='BrBG',center=0)
```

```python
plt.figure(1,figsize=(10,5))
sns.heatmap(corr_matrix,cmap='BrBG',center=0,annot=True)
```

```python
plt.figure(1,figsize=(10,5))
sns.heatmap(corr_matrix,cmap='BrBG',center=0,annot=True,cbar=False,linewidth=0.5,linecolor='purple')
```
